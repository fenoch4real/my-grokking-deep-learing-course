```python
weight = 0.1
lr = 0.01
def neural_network(input, weight):
    prediction = input * weight 
    return prediction

number_of_toes = [8.5] 
win_or_lose_binary = [1] # (won!!!)
input = number_of_toes[0] 
true = win_or_lose_binary[0]
pred = neural_network(input,weight)
error = (pred - true) ** 2 # ** forces a negative error to turn positive
print(error)
```

```python
# Comparing the weight after making prediction with higher weight to see if the error goes upward or downward
lr = 0.1
p_up = neural_network(input,weight+lr)
e_up = (p_up - true) ** 2 
print(e_up)
```

```python
# Comparing the weight after making prediction with lower weight to see if the error goes upward or downward
lr = 0.01
p_dn = neural_network(input,weight-lr)
e_dn = (p_dn - true) ** 2 
print(e_dn)
```

```python
# Compare the errors and learn
if(error > e_dn or error > e_up):
   if(e_dn < e_up):
      weight -= lr
      if(e_up < e_up):
         weight += lr
```

```python
# Simple learning put together
# Hot and cold learning is simple. 
# After making a prediction, you predict two more times, once with a slightly higher weight and again with a slightly lower weight. 
# You then move weight depending on which direction gave a smaller error. 

weight = 0.5
input = 0.5 
goal_prediction = 0.8
step_amount = 0.001 # 0.2 to watch it go back and forth

for iteration in range(1101):
    prediction = input * weight
    error = (prediction - goal_prediction) ** 2
    print("Error:" + str(error) + " Prediction:" + str(prediction))
    
    up_prediction = input * (weight + step_amount) 
    up_error = (goal_prediction - up_prediction) ** 2
    
    down_prediction = input * (weight - step_amount) 
    down_error = (goal_prediction - down_prediction) ** 2
    
    if(down_error < up_error):
        weight = weight - step_amount
    if(down_error > up_error):
        weight = weight + step_amount
```

```python
# Hot and cold is innefficient. You have to predict three time for each weight update.
# Even if you know the direection of the error, you don't know the amount of steps neeecssary to drive it to zero - so to speak

# So let’s measure the error and find the direction and amount!
weight = 0.5
goal_pred = 0.8
input = 0.5

for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2 
    direction_and_amount = (pred - goal_pred) * input # Scaling, negative reversal and stopping. Multipling the pure error with the input = gradint descent which allows you to calculate the amount and direction to change the weigtht to reduce the error.
    weight = weight - direction_and_amount
    print("Error:" + str(error) + " Prediction:" + str(pred))

print("Error:" + str(error) + " Prediction:" + str(pred))
```

What is direction_and_amount?<br>
direction_and_amount represents how you want to change weight. The first part b is what I call pure error, which equals (pred - goal_pred). (More about this shortly.) 
The second part c is the multiplication by the input that performs scaling, negative reversal, and stopping, modifying the pure error so it’s ready to update weight.
Another name for direction and amount is sensitivity.

What is the pure error?<br>
The pure error is (pred - goal_pred), which indicates the raw direction and amount you missed. If this is a positive number, you predicted too high, and vice versa. 
If this is a big number, you missed by a big amount, and so on.
What are scaling, negative reversal, and stopping?
These three attributes have the combined effect of translating the pure error into the absolute amount you want to change weight. 
They do so by addressing three major edge cases where the pure error isn’t sufficient to make a good modification to weight.

What is stopping?<br>
Stopping is the first (and simplest) effect on the pure error caused by multiplying it by input. Imagine plugging a CD player into your stereo. 
If you turned the volume all the way up but the CD player was off, the volume change wouldn’t matter. Stopping addresses this in a neural network. 
If input is 0, then it will force direction_and_amount to also be 0. You don’t learn (change the volume) when input is 0, because there’s nothing to learn. 
Every weight value has the same error, and moving it makes no difference because pred is always 0.

What is negative reversal?<br>
This is probably the most difficult and important effect. Normally (when input is positive), moving weight upward makes the prediction move upward. 
But if input is negative, then all of a sudden weight changes directions! When input is negative, moving weight up makes the prediction go down. 
It’s reversed! How do you address this? Well, multiplying the pure error by input will reverse the sign of direction_and_amount in the event that input is negative. 
This is negative reversal, ensuring that weight moves in the correct direction even if input is negative.

What is scaling?<br>
Scaling is the third effect on the pure error caused by multiplying it by input. Logically, if input is big, your weight update should also be big. This is more of a side effect, because it often goes out of control. Later, you’ll use alpha to address when that happens.

```python
# One iteration of gradient descent to establish what and why we need to change for the the neural network to learn

weight=0.1
alpha = 0.01 # fixed before traing

def neural_network(input, weight):  # 1. An empty network
    prediction = input * weight 
    return prediction

number_of_toes = [8.5] 
win_or_lose_binary = [1] # (won!!!)
input = number_of_toes[0] 
goal_pred = win_or_lose_binary[0]

pred = neural_network(input,weight) 

error = (pred - goal_pred) ** 2     # 2. PREDICT: Making a prediction and evaluating error
delta = pred - goal_pred            # 3. COMPARE: Calculating the node delta and putting it on the output node
weight_delta = input * delta        # 4. LEARN: Calculating the weight delta and putting it on the weight
weight -= weight_delta * alpha      # 5. LEARN: Updating the weight. You multiply weight_delta by small alpha before updating weight to control how fast the network learns. If it learns too fast (jumping about on an error aginst weight ploted graph, where we are look for where slope is equal to zero), it can update weights too aggressively and overshoot.
```

```python
# Observe jump about of the error value for which we typically require alpha to not make it jump about as much aka lear to fast based on updated eight
weight, goal_pred, input = (0.0, 0.8, 1.1)
for iteration in range(4):
    print("-----\nWeight:" + str(weight))
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = delta * input
    weight = weight - weight_delta
    print("Error:" + str(error) + " Prediction:" + str(pred)) 
    print("Delta:" + str(delta) + " Weight Delta:" + str(weight_delta))   
```

```python
# So for the function: error = (pred - goal_pred) ** 2, the goal is to find the error = ((input * weight) - goal_pred) ** 2 while being at liber ty tweak any variable.
# For any input and goal_pred, an exact relationship is defined between error and weight, found by combining the prediction and error formulas. In this case: 
# error = ((0.5 * weight) - 0.8) ** 2. You modify specific parts of an error function until the error value goes to 0. 
# This error function is calculated using a combination of variables, some of which you can change (weights) 
# and some of which you can’t (input data i.e reality of the world you are interested in, output data i.e goal_pred - "setting your goal based on your capability and not what is necessary
# which changing could mean giving up"), and the error logic i.e. a good measure of how much you have missed), likewise you can modify anything in pred function (learning) except the input.
# To sum it all up "learning" is all about automatically changing the prediction function so that it makes good predictions—aka, so that the subsequent error goes down to 0. 
# That is adjust weight for error to become 0.
```

```python
# Now given the relationship between error and weight, HOW can you know can you we use the formular to change weight so error goes in a particular direction.
# Well there is a method for doing this to any formular. 
# It is called DERIVATIVE - however much I mive this part, the other part moves a certain amount as much.
# For example, red_length = blue_length * 2. 2 is the derivative. 
# If derivative is postitive moving one variable moves the other in one direction and if the derivative is negative, moving one variable moves the other in the opposite direction.
# Thus derivate represents direction and amount.
```

```python
# Another ways people look at deriative is a slope at a point of a line or curve.
# Remeber error = ((input * weight) - goal_pred) ** 2 i.e. error = ((0.5 * weight) - 0.8) ** 2
# Ploting it give a quadratic curve (from hot and cold above) and there is a point where error is 0, positive and negative. 
# With derivatives, you can pick any two variables in any formula/function, and know how they interact.
# y = (((beta * gamma) ** 2) + (epsilon + 22 - x)) ** (1/2)
# For any function, you can pick two variables and compute how much one changes when you change the other. 
# 
# Bottom line: you’re going to build neural networks. A neural network is really just one thing: a bunch of weights you use to compute an error function. 
# And for any error function (no matter how complicated), you can compute the relationship between any weight and the final error of the network.
# With this information, you can change each weight in the neural network to reduce error down to 0—and that’s exactly what you’re going to do.
# Derivative aka sensitivity between two variables - positive or negative(opposite movement() or zero (no movement regardless of what you do to another eg y=0*x).
# In our case weight_delta is our derivative. A derivative gives you the relationship between any two variables in a function. 
# You use the derivative to determine the relationship between any weight and error.
# The slope of a line or curve always points in the opposite direction of the lowest point of the line or curve. 
# So, if you have a negative slope, you increase weight to find the minimum of error. Your weight moves in opposite direction of the slope.
# This method for learning (finding error minimums) is called GRADIENT DESCENT! 

weight = 0.0
goal_pred = 0.8
input = 1.1
for iteration in range(4):
    pred = input * weight
    error = (pred - goal_pred) ** 2 
    delta = pred - goal_pred 
    weight_delta = delta * input # Derivative (how fast the error changes, given changes in the weight)
    weight = weight - weight_delta

    print("Error:" + str(error) + " Prediction:" + str(pred))
    
```

```python
# Now that it works, let’s break it. Play around with the starting weight, goal_pred, and input numbers. 
# You can set them all to just about anything, and the neural network will figure out how to predict the output given the input using the weight.
weight = 0.5
goal_pred = 0.8
input = 2 #0.5 - Change input into two we see the prediction jumping around and explode in value (over correcting) which is called divergence. Hence we introduce alpha
for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = input * delta
    weight = weight - weight_delta
    print("Error:" + str(error) + " Prediction:" + str(pred))
```

```python
# Introducing alpha
# What’s the problem you’re trying to solve? That if the input is too big, then the weight update can overcorrect.
# What’s the symptom? That when you overcorrect, the new derivative is even larger in magnitude than when you started (although the sign will be the opposite).
# The solution is to multiply the weight update by a fraction to make it smaller.
# In most cases, this involves multiplying the weight update by a single real-valued number between 0 and 1, known as alpha so it doesn’t overshoot.
# Finding the appropriate alpha, even for state-of-the-art neural networks, is often done by guessing. 
# You watch the error over time. If it starts diverging (going up), then the alpha is too high, and you decrease it.
# If learning is happening too slowly, then the alpha is too low, and you increase it.
# There are other methods than simple gradient descent that attempt to counter for this, but gradient descent is still very popular.

weight = 0.5
goal_pred = 0.8
input = 2
alpha = 0.1

for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2 
    derivative = input * (pred - goal_pred) 
    weight = weight - (alpha * derivative)
    print("Error:" + str(error) + " Prediction:" + str(pred))
```
